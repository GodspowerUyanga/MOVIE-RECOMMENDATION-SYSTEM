# Importing Neccesary Dependencies

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

# <b style = "color:blue"> DATA COLLECTION </b>

### Loading Movie Dataset

```python
movies_data = pd.read_csv("movies.csv")
```

### Let's Peruse The First 5 Rows Of Our Movie Data

```python
movies_data.head()
```

### Know The Shape Of The Movie Data

```python
print(f" We Have {movies_data.shape} Rows And Columns In The Data")
```

#### We Have 4803 Rows And 24 Columns In The Movie Data And We will Extract The Important Futures We Will Use To Build Our System.

### Checking If They Is Any Null Value

```python
movies_data.isnull().sum()
```

### Selecting Important Features For Recommendation

```python
important_features = ['genres','keywords','tagline','cast','director']
print(important_features)
```

### Replacing The Null Values With New Strings

```python
for features in important_features:
    movies_data[features] = movies_data[features].fillna('')
```

### Adding All Our Selected Important Features Together

```python
Added_features = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

### Lets Confirm Our Added Features

```python
print(Added_features)
```

### Check  Our Movie Unique Features

```python
Added_features.unique()
```

### Converting Our Textual Data To Feature Vector 

```python
vectorizer = TfidfVectorizer()
```

### Confirming Our Model

```python
vectorizer
```

### Lets Transform Our Added Feature With Our Vectorizer

```python
feature_vectors = vectorizer.fit_transform(Added_features)
```

```python
print(feature_vectors)
```

#### We have Finally Converted Textual Movie Data To Feature Vector

### Check The Shape Of The Feature Vectors

```python
feature_vectors.shape
```

#### The Feature Vectors Has  4803 Rows And 17318 Columns

## Cosine Similarity Algorithm

### Using Cosine Similarity To Obtain Similarity Scores

```python
similarity_score = cosine_similarity(feature_vectors)
```

```python
print(similarity_score)
```

### Check The Shape Of Similarity Score

```python
similarity_score.shape
```

#### Now The Rows And Columns Of Similarity Score Is Same

### Lets Create a Lists That Contains The Names Of all The Movie Names In The Movie Dataset

```python
List_Movie_Titles = movies_data['title'].tolist()
print(List_Movie_Titles)
```

### Getting Movie Name Input From User

```python
movie_name = input("Enter Your Best Movie : ")
```

### Lets Find The Close Match From Our User Inputed Movie Name

```python
close_match = difflib.get_close_matches(movie_name, List_Movie_Titles)
print(close_match)
```

### Find Close Match Index

```python
close_match_index = close_match[0]
print(close_match_index)
```

### Finding Movie Index With Title

```python
movie_Index = movies_data[movies_data.title == close_match_index]['index'].values[0]
print(movie_Index)
```

### Getting A List Of Similar Movies From User Input

```python
similarity_score_list = list(enumerate(similarity_score[movie_Index]))
print(similarity_score_list)
```

### Lets Check The Size Of Our Similarity Score List

```python
len(similarity_score_list)
```

### Sorting The Movies Based On Their Similarity Score

```python
sorted_similar_movies = sorted (similarity_score_list,  key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

### Printing Similar Movie Name Base On Index

```python
print('HEY HERE IS OUR SUGGESTED MOVIE, YOU WILL LIKE THEM JUST CLICK ON YOUR FAVORITE: \n')

i = 1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    if (i<30):
        print(i, '.',title_from_index)
        i+=1
```

## <b style="color:blue"> MOVIE RECOMMENDATION SYSTEM </b>

```python
User_Name =  input("WHAT IS YOUR NAME : ")

print("-------------------------------------------------------------\n")
movie_name = input("WHAT IS THE NAME OF YOUR DESIRED MOVIE : ")

print("--------------------------------------------------------------\n")

List_Movie_Titles = movies_data['title'].tolist()

close_match = difflib.get_close_matches(movie_name, List_Movie_Titles)

close_match_index = close_match[0]

movie_Index = movies_data[movies_data.title == close_match_index]['index'].values[0]

similarity_score_list = list(enumerate(similarity_score[movie_Index]))

sorted_similar_movies = sorted (similarity_score_list,  key = lambda x:x[1], reverse = True)

print(f'{User_Name} CHECK RELATED MOVIES BELOW : \n')

i = 1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    if (i<30):
        print(i, '.',title_from_index)
        i+=1
```
